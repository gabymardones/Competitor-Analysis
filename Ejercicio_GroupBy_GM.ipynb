```python
import pandas as pd

df= pd.read_csv('/content/CalIt2 (2).csv', header = None)
df.head()

```

```python
df.rename(columns = {0:"flow_ID", 1:"Date", 2:"Time", 3:"Count" }, inplace=True)
df

```

```python
df.info()
```

```python
filtro = (df["Date"] == "07/24/05") & (df["flow_ID"] == 7)
datos_filtrados = df[filtro]
datos_filtrados
```

```python
datos_filtrados.sort_values(by ="Count", ascending=False)
```

```python
datos_ordenados = datos_filtrados.sort_values("Count", ascending=False)
Top_10 = datos_ordenados.head(10)
```

```python
menor =Top_10 ["Count"].min()
#Apply
Top_10["Count"].apply(lambda x:x-menor).mean()
```

```python
df["Date"] = pd.to_datetime(df["Date"], format="%m/%d/%y")
```

```python
filtro =(df["Date"].dt.month == 8) & (df["Date"].dt.year == 2005) & (df["flow_ID"] == 7)
datos_agosto = df[filtro]
datos_agosto.groupby(by="Date")["Count"].max()


```
